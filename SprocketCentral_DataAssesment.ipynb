# <U> 3. INFERENCES </U>
### <U > QUALITY ISSUES </U>
##### `TRANSACTIONS` table
- There are totally 555 rows with atleast one missing value.
- The column 'online_order' contains the most number(360) of missing values followed by other columns(each of these columns having 197 missing values)
  >brand                    
  >product_class              
  >product_size 
  >product_line                
  >standard_cost                
  >product_first_sold_date
- Most of these missing values correspond to the product_id '0'
- The column "product_first_sold_date" datatype has to be changed from float to datetime for analytical purposes  

##### `CUSTOMER DEMOGRAPHIC` table
- Lots of rows with missing values are found. 
- Some column names could be altered into a shorter form for analytical purposes.
- The column 'gender' has Male,Female values and their abbreviations other times.(There are also inconsistent values with spelling errors)
- There is an outlier found in the DOB column that corresponds to an unreasonable age.
- There is an unwanted column named 'default' filled with gibberish and could be removed for better data exploration.

##### `CUSTOMER ADDRESS` table
- Erroneous datatype for the postcode column (It is a string and not an integer)
- Unnecessary zeroes are present in the address column.
- The state column is filled with the abbreviation of the the respective states but sometimes it includes the full-form as well.(Inconsistent value)

# 4. CLEANING 

```python
transac_clean = transac.copy()
cusdem_clean = cusdem.copy()
cusadd_clean = cusadd.copy()
```

`TRANSACTIONS TABLE`

```python
transac_clean.tail()
```

```python
transac_clean.dropna(how ='any',inplace = True)
```

```python
transac_clean.info()
```

```python
final_date = []
```

```python
import datetime
serial = transac_clean['product_first_sold_date'].tolist()
for i in range(len(serial)):
    seconds = (serial[i] - 25569) * 86400.0
    finaldate = datetime.datetime.utcfromtimestamp(seconds)
    final_date.append(finaldate)
```

```python
transac_clean['sold_on'] = final_date 
```

```python
transac_clean.drop(columns =['product_first_sold_date'],inplace = True)
```

```python
transac_clean.head(3)
```

```python
transac_clean.info()
```

`CUSTOMER DEMOGRAPHIC`

```python
cusdem_clean.dropna(how='any',inplace= True)
```

```python
cusdem_clean.head()
```

```python
cusdem_clean.info()
```

```python
cusdem['gender'].value_counts()
```

```python
cusdem_clean['gender'].value_counts()

gender_abbrev = {'M': 'Male',
                'F': 'Female',
                'Femal': 'Female',
                'U':'Unknown' }
print(gender_abbrev)

cusdem_clean.replace({"gender": gender_abbrev},inplace = True) 
```

```python
cusdem_clean['gender'].value_counts()
```

```python
from datetime import *

now = pd.to_datetime('now')
cusdem_clean['age'] = (now.year - cusdem_clean['DOB'].dt.year) - ((now.month - cusdem_clean['DOB'].dt.month) < 0)
```

```python
cusdem_clean['age'].sort_values(ascending=False)
```

```python
cusdem_clean.drop(columns =['DOB','default'],inplace = True)
```

```python
cusdem_clean.tail()
```

```python
cusdem_clean.rename(columns = {'past_3_years_bike_related_purchases': '3y_bikes_purchased'},inplace = True)
```

```python
cusdem_clean.rename(columns = {'deceased_indicator': 'D_indicator'},inplace = True)
```
